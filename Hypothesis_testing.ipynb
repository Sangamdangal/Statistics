```python
import pandas as pd
from scipy import stats
```

```python
education_districtwise = pd.read_csv('C:\\Users\\hp\\Desktop\\PYTHON\\Stat\\Day 3\\education_districtwise.csv')
education_districtwise = education_districtwise.dropna()    #used to remove missing values from a DataFrame
```

```python
education_districtwise
```

```python
state21 = education_districtwise[education_districtwise['STATCD']== 21]
state21
```

```python
state28 = education_districtwise[education_districtwise['STATCD']== 28]
state28
```

```python
#sample size n=20
#choose True cuz sampling with replacement replace=True
#choose arbitary number for random seed random_state=13450

sampled_state21 = state21.sample( n=20, replace=True, random_state=200)
sampled_state21
```

```python
sampled_state28 = state28.sample( n=20, replace=True, random_state=210)
sampled_state28
```

```python
sampled_state21['OVERALL_LI'].mean()
```

```python
sampled_state28['OVERALL_LI'].mean()
```

CONDUCT HYPOTHESIS TEST

1. state null and alternative hypothesis
2. choose significance level
3. find p-value
4. reject or fail to reject null hypothesis

```python
stats.ttest_ind(a= sampled_state21['OVERALL_LI'], b= sampled_state28['OVERALL_LI'], equal_var=False)   #equal_var set to false cuz we dont want to assume two sample having same variance
```

```python
#t value is statistic=0.2664...... 
#p value 0.792... 79.2% which is more than 5% is null hypothesis
```
