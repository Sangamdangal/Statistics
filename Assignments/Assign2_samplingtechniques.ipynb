• Step 1: Since analyzing the entire dataset is computationally expensive, you need to use
sampling techniques. Implement the following sampling methods:


• Simple Random Sampling: Randomly select 10% of the customers from the
dataset.

• Stratified Sampling: Select a sample based on Purchase_Category, ensuring
each category is proportionally represented.

• Systematic Sampling: Select every 10th customer in the dataset for analysis.

• Cluster Sampling: Group customers by country and select a few countries to
analyze all customers within those countries.

• Step 2: For each sampling method, perform the following:

• Compare the mean and standard deviation of Purchase_Amount with the entire

• Discuss whether the sample is representative of the overall population.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset

ecommerce_data = pd.read_csv('ecommerce_data.csv')
ecommerce_data.head()
```

```python
# Simple Random Sampling: Randomly select 10% of the customers from the dataset.

sample_fraction = 0.10    # 10%
simple_random_sample = ecommerce_data.sample(frac=sample_fraction, random_state=1)
simple_random_sample.head()
```

```python
# Select a sample based on Purchase_Category, ensuring each category is proportionally represented.

stratified_sample = ecommerce_data.groupby('Purchase_Category').apply(lambda x: x.sample(frac=sample_fraction, random_state=1)).reset_index(drop=True)
stratified_sample.head()
```

```python
# Select every 10th customer in the dataset for analysis.#systematic_sample

systematic_sample = ecommerce_data.iloc[::10]
systematic_sample.head()
```

```python
# Discuss whether the sample is representative of the overall population.

#Simple Random Sampling and Stratified Sampling are generally effective in providing samples that are representative of the overall population.

#Systematic Sampling is often representative, but its effectiveness depends on the absence of cyclical patterns that align with the sampling interval.

#Cluster Sampling’s representativeness is contingent upon the diversity and coverage of the selected clusters. If the clusters are well-chosen to represent the population's diversity, the sample will be representative; otherwise, it may be biased towards the selected clusters.
```

```python
# Plot the distribution of Purchase_Amount for each sample.

# Set up the matplotlib figure
plt.figure(figsize=(16, 12))

# Plot Simple Random Sampling
plt.subplot(2, 2, 1)
sns.histplot(simple_random_sample['Purchase_Amount'], kde=True, color='blue', bins=30)
plt.title('Distribution of Purchase_Amount - Simple Random Sampling')

# Plot Stratified Sampling
plt.subplot(2, 2, 2)
sns.histplot(stratified_sample['Purchase_Amount'], kde=True, color='green', bins=30)
plt.title('Distribution of Purchase_Amount - Stratified Sampling')

# Plot Systematic Sampling
plt.subplot(2, 2, 3)
sns.histplot(systematic_sample['Purchase_Amount'], kde=True, color='orange', bins=30)
plt.title('Distribution of Purchase_Amount - Systematic Sampling')

# Plot Cluster Sampling
plt.subplot(2, 2, 4)
sns.histplot(cluster_sample['Purchase_Amount'], kde=True, color='red', bins=30)
plt.title('Distribution of Purchase_Amount - Cluster Sampling')

# Adjust layout
plt.tight_layout()
plt.show()
```
