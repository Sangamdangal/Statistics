STATISTICAL TEST

HYPOTHESIS 1
ROA is considering a metropolitan-focused approach. Within
California, they want to know if the mean AQI in Ohio County is statistically different from the rest of California.

```python
aqi_california = aqi[aqi['state_name'] == 'California']
aqi_california.head(5)
```

```python
aqi_new_york = aqi[aqi['county_name'] == 'New York']
aqi_new_york
```

```python
sample_aqi_california = aqi_california.sample( n=20, replace=True, random_state=100)
sample_aqi_california.head(10)
```

```python
sample_aqi_new_york = aqi_new_york.sample( n=20, replace=True, random_state=100)
sample_aqi_new_york.head(5)
```

```python
stats.ttest_ind(a= sample_aqi_california['aqi'], b= sample_aqi_new_york['aqi'], equal_var=False)   #equal_var set to false cuz we dont want to assume two sample having same variance
```

```python
p_value = 1.5174361140551958e-06
alpha = 0.5
if p_value < alpha:
    print("The mean AQI in California is statistically different from the rest of New York.")
else:
    print("There is not enough evidence to say the mean AQI in California is different from the rest of New York.")

#p value 1.517..-06 which is less than 0.5 so we reject null hypothesis
```

Hypothesis 2: With limited resources, ROA has to choose between New York
and Ohio for their next regional office. Does New York have a lower AQI than

```python
aqi_arizona = aqi[aqi['state_name'] == 'Arizona']
aqi_arizona.head(5)
```

```python
sample_aqi_ohio['aqi'].mean()
```

```python
#Ohio has lower AQI than Arizona.
```

```python
stats.ttest_ind(a= sample_aqi_arizona['aqi'], b=sample_aqi_ohio['aqi'])
```

Hypothesis 3: A new policy will affect those states with a mean AQI of 10 or
greater. Will Michigan be affected by this new policy?


```python
aqi_michigan = aqi[aqi['state_name'] == 'Michigan']
aqi_michigan.head()
```

```python
sample_aqi_michigan = aqi_michigan.sample(n=20, replace= True, random_state= 1200)
sample_aqi_michigan.head()
```

```python
sample_aqi_michigan['aqi'].mean()
```

```python
stats.ttest_ind(sample_aqi_michigan['aqi'],8.6,alternative='greater')
```

p_value is equal to 0.5 which is alternative hypothesis . This suggests there is not enough evidence to conclude that the mean AQI in Michigan is greater than 10. Thus, based on this test, Michigan would not be affected by the new policy
